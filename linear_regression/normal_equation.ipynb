```python
x_train=np.array([1,2,3,4,5])
y_train=np.array([2,4,6,8,10])
```

```python
X = np.column_stack((np.ones(len(x_train)), x_train))
X
```

```python
w=np.linalg.pinv(X)@y_train
```

```python
print("bias,weight",w)
```

```python
residual=y_train-X@w
```

```python
print("orthogonality chk",X.T@residual)
```

```python
print("residual norm",np.linalg.norm(residual))
```

```python
x_test=np.array([10,20])
x_test=np.array(np.column_stack((np.stack(np.ones(len(x_test))),x_test)))
```

```python
y_test=x_test@w
```

```python
print("result",y_test)
```

```python
y_pred=X@w
plt.scatter(x_train, y_train, label="Actual data")
plt.plot(x_train, y_pred, label="Best fit line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
