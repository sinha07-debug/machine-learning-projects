```python
import numpy as np
```

```python
def matrix_design(X):
    return np.column_stack((np.ones(len(X)),X))
```

```python
def loss_fn(X,y,w):
    n=len(y)
    error=X@w-y
    return (1/n)*np.dot(error,error)
```

```python
def gradient(X,y,w):
    n=len(y)
    return (2/n)*(X.T)@(X@w-y)
```

```python
def gradient_descent(X,y):
    lr=0.01
    epoch=1000
    w=np.zeros(X.shape[1])
    for _ in range(epoch):
        w-=lr*gradient(X,y,w)
    return w
    
```

```python
X_train=np.array([1,1.5,2,2.3,3,3.6,4,5])
Y_train=np.array([3,2,5,7,7.5,8.9,9,11])
```

```python
X=matrix_design(X_train)
```

```python
w_gd=gradient_descent(X,Y_train)
```

```python
print("bias:",w_gd[0])
print("weight:",w_gd[1])
print("loss total:",loss_fn(X,Y_train,w_gd))
```

```python
import matplotlib.pyplot as plt

# predictions
y_pred = X @ w_gd

# scatter plot of actual data
plt.scatter(X_train, Y_train, label="Actual data")

# fitted line
plt.plot(X_train, y_pred, label="Fitted line")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
